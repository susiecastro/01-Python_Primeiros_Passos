#O que são variáveis?

Variáveis são um dos recursos mais básicos das linguagens de programação.

Através da utilização de variáveis, podemos armazenar dados na memória do computador para acessá-los mais tarde, quando forem necessários.

#Declaração e atribuição

Algumas regras devem ser seguidas ao criarmos um nome para uma variável:


*   precisa começar com uma letra ou um underscore
*   não pode começar com número
*   não pode conter caracteres especiais (*,$,%,#,@,...etc)
*   nomes de variáveis são cases sensitves (nome é diferente de Nome)
*   não pode ser uma palavra reservada da linguagem
