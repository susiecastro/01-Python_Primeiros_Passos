#Estrutura de repetição While

Enquanto na estrutura **for** nós temos a quantidade de repetições já definidas, no while elas acontecem até que uma condição de parada seja atingida

```python
#A primeira coisa é criar uma varíavel para controlar as repeticões. Exemplo
contador = 0
```

```python
while contador <10:
  print(contador)
  contador = contador +1
```

```python
bombons = 10
```

```python
while bombons >0:
  print (f'Eu tenho {bombons} bombons!')
  bombons = bombons -1
  print (f'Comi 1 e fiquei com {bombons}')
```
