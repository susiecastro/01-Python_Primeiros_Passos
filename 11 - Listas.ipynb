#Listas

Uma lista é uma estrutura de dados composta por itens organizados de forma linear, na qual cada um pode ser acessado a partir de um indice, que representa sua posição na coleção (iniciando em zero)

  *são mutáveis
  *podem conter tipos diferentes de objetos

##Criando uma lista

```python
#Uma lista é criada dentro de colchetes. Exemplo:
lista1=['Maura','Deise','Susie']
```

```python
print(lista1)
```

```python
lista2 = [10,True,'Susie',125.4,'Python']
```

```python
print((lista2))
```

```python
lista3 =[10,'Susie',[1,2,3]]
```

##Acessando elementos de uma lista pelo indice

```python
print(lista2)
```

```python
#Acessando indices da lista, para acessar primeiro, segundo, terceiro....ou qualquer elemento
print(lista2[0])
```

```python
print(lista2[2])
```

```python
#Acessando o último elemento da lista
print(lista2[-1])
```

##Manipulando elementos de uma lista

###Adicionando elementos

#####Método append()

O método append() insere um elemento sempre no final da lista

```python
lista2
```

```python
lista2.append('SEXTOOOOU')
```

####Método insert()

O método insert() recebe dois argumentos: o índice onde vamos inserir o elemento e o valor do elemento

```python
#No insert() é passada a posição onde o item será inserido e o item que será inserido, respectivamente
lista2.insert(1,'banana')
```

```python
lista2
```

####Excluindo elementos

#####Método pop()

O método pop() exclui o último elemento de uma lista

```python
lista2
```

```python
#Comando sempre exclui o último comando da lista
lista2.pop()
```

```python
lista2
```

#####Método remove()

O método remove() permite especificar qual elemento deverá ser removido

```python
#Para utilziar o comando basta digitar o valor que você não deseja mais na lista.
lista2.remove('banana')
```

```python
lista2
```

#####Comando del

Podemos excluir um elemento de uma lista utilizando o comando del seguido do nome de lista e o índice do elemento que deverá ser excluído

```python
lista2
```

```python
#Para usar o comando del, digite del o nome da lista e o indice do elemento que deseja deletar
del lista2[2]
```

```python
lista2
```
