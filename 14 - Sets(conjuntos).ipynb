#Sets (conjuntos)

São esttruturas de dados destinadas a armazenar um conjunto de dados distintos. Suas principais características são:

* set = {elemento1, elmento2, elmento3...}
* elmentos são desordenados
* elementos não possuem índices
* não podem conter elmentos
* permitem operações de conjuntos
* são mutáveis

#Criando um conjunto

```python
conjunto1={1,2,3,4,5}
```

```python
print(conjunto1)
```

```python
type(conjunto1)
```

```python
conjunto2={1,2,2,2,3,4,5,6}
```

```python
#Elementos repetidos não aparecem no conjunto
print(conjunto2)
```

#Algumas operações de conjuntos

###União

```python
conjunto3 ={1,2,3}
conjunto4 ={4,2,5}
```

```python
#A união segue o conceito matemático
conjunto3.union(conjunto4)
```

###Interseção

```python
#Interseção mostra o elemento comum entre os conjuntos
conjunto3.intersection(conjunto4)
```
