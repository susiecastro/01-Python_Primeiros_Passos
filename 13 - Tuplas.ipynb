#Tuplas

Tuplas são uma estrutura de dados do Python, cujo objetivo (assim como as listas) é armazenar diversos dados. Algumas características das tuplas são:

* tupla = (elemento1, elemento2, elemento3...
* é uma sequência ordenada de elementos
* seus elementos possuem índice
* podem ser heterôgeneas (possuirem diferentes tipos de dados)
* é **imutável**

##Criando uma tupla

```python
numeros =(1,2,3,4,5,6,7,8,9)
```

```python
numeros
```

```python
type(numeros)
```

```python
pessoas=('Maria',34,'João',27,'Gabriel',14,'Mariana',41)
```

```python
print(pessoas)
```

```python
type(pessoas)
```

```python
#Acessar os dados de uma tupla através do número do indice
pessoas[0]
```

```python
pessoas[4]
```

```python
pessoas[-1]
```

##Tuplas são imutáveis

Logo, não conseguimos manipular os elementos de uma tupla

```python
pessoas
```

```python
#Como as tuplas são imutáveis, qualquer tentativa de mudança dará erro
pessoas[0]='Joana'
```

```python
del pessoas[2]
```
