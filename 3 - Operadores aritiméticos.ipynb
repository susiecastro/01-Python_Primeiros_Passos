#Operadores aritméticos

Todo mundo já usou operadores aritméticos na escola! Nos primeiros anos aprendemos a fazer continhas de soma, subtração, multiplicação e divisão.

No Python também temos esses operadores e podemos realizar todos estes cálculos

##Adição

```python
print(10+20)
print(30+40)
print(10+70+90+1100+10000)
```

##Subtração

```python
print(123-87)
print(67-198)
print (64-8-78-90-167)
```

##Multiplicação

```python
print(10*8)
```

```python
print(12*3*5)
```

```python
print(12*3*5)
```

##Divisão

```python
print(100/10)
print(129/23)
#sempre o resultado da divisão sairá com casa decimal
```

```python
print(int(100/10))
#caso queira que o resultado da divisão saia sem casa decimal colocar a divisão dentro da fiunção int()
```

###Divisão com o resultado inteiro

```python
print(129//23)
```

```python
print(1345//57)
```

###Módulo

```python
print (129%23)
#resto da divisão
```

```python
print(10%3)
```

##Exponenciação

```python
print(10**2)
print(10**4)
print(3**3)
```
