#Tipagem Dinâmica

Tipagem Dinâmica é uma característica de determinadas linguagens de programação, que não exigem declarações de tipo de dados, pois são capazes de escolher que tipo utilizar dinamicamente para cada variável, podendo alterá-lo durante a compilação ou a execução do programa.

Python é uma linguagem de tipagem forte e dinâmica, ou seja, não é preciso em momento algum definir o tipo de dado que uma variável irá receber

```python
nome_professor ="Susie"
```

```python
print(nome_professor)
type(nome_professor)
```

**Muda o tipo de dados quando o valor inputado na variável é alterado**

```python
nome_professor=703
```

```python
print(nome_professor)
type(nome_professor)
```

```python
nome_professor = True
```

```python
print(nome_professor)
type(nome_professor)
```
