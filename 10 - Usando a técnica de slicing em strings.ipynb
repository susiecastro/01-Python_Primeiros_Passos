#Slicing (fatiamento)

A técnica de fatiamento nos permite acessar partes de uma string utilizando a notação de índice

#Acessando o índice

```python
frase = "Meu nome é Susie e eu estou estudando Python"
```

```python
frase[0]
```

```python
#vale lembrar que o espaço deve ser contato para contar o índice
frase[11]
```

#Notação de dois pontos

```python
#pegar dois indíces o primeiro índice é inclusivo e o último exclusivo.Exemplo [0:3] pegará os caracteres do indíce 0 até o 2
frase[0:3]
```

```python
frase[11:16]
```

```python
#Caso necessário pegar as informações de uma frase de um ponto até o final. Exemplo [11:)]
frase[11:]
```

```python
frase[:]
```

```python
print(frase)
```

```python
#Slicing de traz para frente
frase[-7:]
```

#Steps

```python
print(frase)
```

```python
frase[0:3]
```

```python
frase[0:16]
```

```python
#Você pode colocar o slice e pular de letras em letras, Exemplo [0:17:4] - considerará do índice 0 até o 16 e irá pular de 4 em 4
frase[0:17:4]
```

#Invertendo uma string com slicing

```python
frase[:]
```

```python
#invertendo uma string
frase[::-1]
```
