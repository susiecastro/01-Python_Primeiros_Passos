#Dicionários

O dicionário é uma estrutura de dados em Python utilizada para armazenar várias informações. Os dicionários são organizados em pares de chaves e valores

* Notação (chave: valor)
* São mutáveis
* Chaves precisam ser únicas

##Criando um dicionário

```python
dados = {'nome': 'Susie',
         'estado': 'São Paulo',
         'linguagem': 'Python'}
```

```python
print(dados)
```

```python
dados1 = {'nome': 'Susie',
         'nome': 'Fernando',
         'estado': 'São Paulo',
         'linguagem': 'Python'}
```

```python
# Apareceu somente um nome porque as chaves precisam ser únicas, caso a chave tenha nomes repetidos o Python sobrescreve o útlimo valor
print(dados1)
```

```python
dados3={'nome':'Mariana',
        'idade': 27,
        'estado':'Minas Gerais',
        'cidade':'Belo Horizonte'}
```

```python
print(dados3)
```

```python
dados_numericos={1:'número um',
                 2:'número dois',
                 3:'número três'}
```

```python
print(dados_numericos)
```

```python
type(dados1)
```

```python
type(dados3)
```

```python
type(dados_numericos)
```
