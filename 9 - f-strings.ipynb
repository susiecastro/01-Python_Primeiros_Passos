#f-strings

Também chamadas de **strings literais formadas** (formated strings literals), f-strings são strings com a letra f no início e chaves {} para realizar a interpolação de variáveis ou expressões

```python
idade= 35
```

```python
print(idade)
```

```python
print (f'Meu nome é Susie e eu tenho {idade} anos')
```

```python
print (f'Meu nome é Susie e eu tenho {30+8} anos')
```

```python
linguagem ='Python'
```

```python
print(f'Eu estou aprendendo a programar em {linguagem}!')
```

```python
linguagem ='C#'
```

```python
print(f'Eu estou aprendendo a programar em {linguagem}!')
```
