#Funções
São blocos de códigos reutilizáveis que executam uma determinada tarefa

*podem ser definidas pelo programador

*podem ser built-in (internas Python)

*podem ser de terceiros (pacotes)

*são criadas com a palavra reservada def (definition)

#Criando uma função simples

```python
#imagine que em vários espaços do seu código vc precisa imprimir estas frases para o usuário
print('Seja bem vindo(a) à EmpowerData')
print('Você está logado(a) no sismtema')
print('Não se esqueça de fazer o logout!')
```

```python
#criando uma função dos três prints acima
def imprimir_frases ():
  print('Seja bem vindo(a) à EmpowerPython')
  print('Você está logado(a) no sismtema')
  print('Não se esqueça de fazer o logout!')
```

```python
imprimir_frases()
```

```python
imprimir_frases()
```

#Criando uma função com parâmetros

Parâmetros são valores que podemos passar para uma função

```python
print('Seja bem vindo(a) à EmpowerData, Susie')
print('Susie, você está logado(a) no sismtema')
print('Não se esqueça de fazer o logout Susie!')
```

```python
def imprimir_frases_usuario(nome):
  print(f'Seja bem vindo(a) à EmpowerData, {nome}')
  print(f'{nome}, você está logado(a) no sistema')
  print(f'Não se esqueça de fazer o logout {nome}')
```

```python
imprimir_frases_usuario('Susie')
```

```python
imprimir_frases_usuario('Deise')
```

```python
#Exemplo de uma função criada para somar dois números
def somar_numeros(numero1,numero2):
  print(numero1+numero2)
```

```python
somar_numeros(30,40)
```

#Escopo de variáveis em funções

Variáveis declaradas dentro de uma função só são visíveis dentro do bloco da função

```python
#Variáveis criadas dentro de uma função só podem ser visualizadas dentro da função - escopo local
def imprimir_frases_usuario1(nome):

  nome_usuario = nome + '!'

  print(f'Seja bem vindo(a) à EmpowerData, {nome_usuario}')
  print(f'{nome_usuario}, você está logado(a) no sistema')
  print(f'Não se esqueça de fazer o logout {nome_usuario}')

imprimir_frases_usuario1('Susie')
```

```python
#Variavel não acessível porque foi criada dentro da função
print(nome_usuario)
```

```python
#Criando variável fora do bloco da função - escopo global

nome_linguagem = 'Python'

def imprimir_frases_usuario1(nome):

  nome_usuario = nome + '!'

  print(f'Seja bem vindo(a) à EmpowerData, {nome_usuario}')
  print(f'{nome_usuario}, você está logado(a) no sistema')
  print(f'Não se esqueça de fazer o logout {nome_usuario}')

imprimir_frases_usuario1('Susie')
```

```python
print(nome_linguagem)
```

#Retorno

Funções podem retornar um valor. Para isso utilizamos a palavra reservada return

```python
def somar_numeros(numero1,numero2):
  return numero1+numero2
```

```python
somar_numeros(10,20)
```

```python
resultado =somar_numeros(300,100)
```

```python
print(resultado)
```

```python
def somar_numeros(numero1,numero2):
  resultado1 = numero1 + numero2
  return resultado1
```

```python
somar_numeros(1000,100)
```

```python
def calculos (numero1,numero2):
  soma = numero1 + numero2
  multiplicacao = numero1*numero2
  exponenciacao = numero1 **numero2

  return soma, multiplicacao, exponenciacao
```

```python
calculos (10,2)
```

```python
calculos (10,3)
```

```python
resultados2 = calculos(10,3)
```

```python
print(resultados2)
```

```python
type(resultados2)
```
