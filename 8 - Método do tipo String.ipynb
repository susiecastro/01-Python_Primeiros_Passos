#Método do tipo String

##split()
Divide uma string, convertendo-a em uma lista

```python
frase ='Meu nome é Susie. Eu estou estudando a linguagem Python!'
```

```python
print(frase)
```

```python
#o split faz uma lista com a frase que foi feita
print(frase.split())
```

```python
#você escolhe um caractere na frase e o mesmo será utilizado como limitador
print(frase.split('.'))
```

```python
'Meu nome é Susie. Eu estou estudando a linguagem Python!'.split()
```

```python
type('Meu nome é Susie. Eu estou estudando a linguagem Python!'.split())
```

##strip()

Remove espaços em branco no início e do fim de uma frase

```python
frase= '   Meu nome é Susie   '
```

```python
print(frase.strip())
```

##capitalize()
Retorna uma string com a primeira letra maíuscula e as demais minúsculas

```python
frase ='meu nome é Susie e eu gosto muito de estudar Python'
```

```python
print(frase.capitalize())
```

##lower()
Retorna uma lista com todas as letras minúsculas

```python
print(frase.lower())
```

##upper()

Retorna uma lista com todas as letras maiusculas

```python
print(frase.upper())
```

##title()

Retorna uma string onde a primeira letra de cada palavra será maiuscula

```python
print(frase.title())
```
