#Desafio

Neste desafio, vamos aumentar um pouquinho do nível de complexidade do exercício. Vamos utilizar conceitos que praticamos acima, junto com as estruturas de repetição e condicionais para resolver um problema.

**Problema**

A lista abaixo contém as seguintes informações de pessoas que pretendem dirigir:

* nome: nome da pessoa
* carteira: **True** para as pessoas que possuem **False** para as que não possuem
* idade: idade da pessoa

Precisamos percorrer toda a lista a verificar se a pessoa pode ou não dirigir. Só podem dirigir as que possuem carteira E que a idade for maior ou igual a 18 anos.

* para as pessoas que podem dirigir, imprima a frase: **nome, pode dirigir**
* para as que não podem, imprima: **nome não pode dirigir**



```python
pessoas=[('Vinicius',True,21),('Gabriel',False,25),('Ana',True,20),('Victor',True,30),('Lucas',False,15)]
```

```python
print(pessoas)
```

```python
for item in pessoas:
  nome_pessoas= item[0]
  if item [1]==True and item[2]>=18:
    print(f'{nome_pessoas}'', ''pode dirirgir')
  else:
    print(f'{nome_pessoas}'', ''não pode dirigir')
```
